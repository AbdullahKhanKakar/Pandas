```python
import pandas as pd
```

```python
df = pd.read_csv("temperature.csv")
```

```python
df
```

## Group by

```python
group = df.groupby("city")
group
```

#### Now data is stored in key value pairs , here city names are the keys

<!-- ![Picture of split apply combine](Capture2.png) -->
<p align="center"></p>

<img src="Capture.png" alt="group_by">

```python
for city,city_data in group:
    print(f"City: {city}")
    print(city_data)
```

```python
group.get_group("paris")
```

```python
new_group = df.groupby(["city"])[["temperature","windspeed"]].agg(["mean","sum","count"])
new_group
```
