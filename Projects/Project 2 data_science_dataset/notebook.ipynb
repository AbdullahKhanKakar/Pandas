<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;">Clean: Salary Estimate Column</h1>

```python
<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;"></h1>
# df["Salary Estimate"] = [entry.replace for entry in df["Salary Estimate"]]
```

```python
import re

def extract_salary_range(data):
    pattern = re.search(r'\$(\d+)K-\$(\d+)K ',data)
    
    if pattern:
        Min = int(pattern.group(1))
        Max = int(pattern.group(2))
        Range = f"{Min}-{Max}"
        return Range
    else:
        return 0        

df["Salary Estimate"] = df["Salary Estimate"].apply(extract_salary_range)
```

```python
df.head(50)
```

<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;">Clean: Revenue Column</h1>

```python
df["Revenue"] = [entry.replace("Unknown / Non-Applicable","$0 to $0 million (USD)") for entry in df["Revenue"]]
```

```python
df.head(1)
```

```python
def extract_revenue_range(data):
    pattern1 = re.search(r'\$(\d+)\+',data)
    pattern2 = re.search(r'\$(\d+) to \$(\d+)',data)
    pattern2_similar = re.search(r'\$(\d+) million to \$(\d+)',data)
    
    if pattern1:
        revenue = int(pattern1.group(1))
        revenue_ = f"{revenue}+ Million"
        return revenue_
    elif pattern2:
        min_revenue = int(pattern2.group(1))
        max_revenue = int(pattern2.group(2))
        revenue_ = f"{min_revenue}-{max_revenue} Million"
        return revenue_
    elif pattern2_similar:
        min_revenue = int(pattern2_similar.group(1))
        max_revenue = int(pattern2_similar.group(2))
        revenue_ = f"{min_revenue}-{max_revenue} Million"
        return revenue_
    else:
        return 0

# df["Revenue"] = df["Revenue"].astype(str)
df["Revenue"] = df["Revenue"].apply(extract_revenue_range)
```

```python
df.head(60)
```

<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;">Clean: Size Column</h1>

```python
def extract_size(data):
    pattern1 = re.search(r"(\d+) to (\d+) employees",data)
    pattern2 = re.search(r"(\d+)\+ employees",data)
    if pattern1:
        Min_size = int(pattern1.group(1))
        Max_size = int(pattern1.group(2))
        size_ = f"{Min_size}-{Max_size}"
        return size_
    elif pattern2:
        value = int(pattern2.group(1))
        size_ = f"{value}+"
        return size_
    else:
        return 0

df["Size"] = df["Size"].apply(extract_size)
```

```python
df.head(52)
```

<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;">Clean: Competitors Column</h1>

```python
df["Competitors"] = [entry.replace("-1","No any competition") for entry in df["Competitors"]]
```

<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;">Add: Min Salary and Max Salary Column</h1>

```python
df.head(2)
```

```python
df["Minimum_Salary"] = df["Salary Estimate"].str.extract(r'(\d+)-')
df["Maximum_Salary"] = df["Salary Estimate"].str.extract(r'-(\d+)')
```

```python
df.head(50)
```

<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;">Add: Average Salary Column</h1>


```python
def average_salary(data):
    match = re.search(r'(\d+)-(\d+)', data)
    if match:
        min_salary = int(match.group(1))
        max_salary = int(match.group(2))
        return (min_salary + max_salary) / 2
    else:
        return 0
        
df["Salary Estimate"] = df["Salary Estimate"].astype(str)
df["Average_Salary"] = df["Salary Estimate"].apply(average_salary)
```

```python
df.head(50)
```

<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;">Add: Location Site Column</h1>

```python
df["Location_State"] = df["Location"].str.extract(r',\ ?([a-zA-Z]+)')
```

```python
df.head(50)
```

<h1 style="background-color:#03398f;padding:10px 0 10px 10px;display:inline;color:white;">Add: Seniorship Column</h1>

```python
def find_senior(data):
    pattern = re.search(r'Sr|Senior',data)
    if pattern:
#         value = pattern
        return "Yes"
    else:
        return "no"

df["Seniorship_required"] = df["Job Title"].apply(find_senior)
```
