<h1 style="background-color:green;padding:10px 0 10px 10px;display:inline;color:white;">Clean author,narrative,price columns</h1>

```python
df["author"] = [entry.replace("Writtenby:", "") for entry in df["author"]]
df["narrator"] = [entry.replace("Narratedby:", "") for entry in df["narrator"]]
df["price"] = [entry.replace(",","") for entry in df["price"]]
```

```python
df.head(2)
```

<h1 style="background-color:green;padding:10px 0 10px 10px;display:inline;color:white;">Clean stars column and convert into stars and ratings</h1>

```python
df["stars"] = [entry.replace("Not rated yet","0 out of 5 stars0 ratings") for entry in df["stars"]]
```

```python
df['Stars'] = df['stars'].str.extract(r'([\d.]+) out of 5 stars')
df['Ratings'] = df['stars'].str.extract(r'(\d+) ratings')
```

```python
df.drop("stars",axis=1,inplace=True)
```

<h1 style="background-color:green;padding:10px 0 10px 10px;display:inline;color:white;">Clean Time column</h1>

```python
import pandas as pd
import re

# Define a function to extract minutes from a string
def extract_minutes(time_str):
    # Use regular expressions to extract hours and minutes
    pattern = re.search(r'(\d+) hrs and (\d+) mins',time_str)    
    hours_match = re.search(r'(\d+) hrs', time_str)
    minutes_match = re.search(r'(\d+) mins', time_str)

    total_minutes = 0
    
    if pattern:
        hours = int(pattern.group(1))
        minutes = int(pattern.group(2))
        total_minutes = (hours * 60) + minutes
        return total_minutes
    elif hours_match:
        total_minutes += int(hours_match.group(1)) * 60
        return total_minutes
    elif minutes_match:
        total_minutes += int(minutes_match.group(1))
        return total_minutes
    else:
        # If the format doesn't match, consider it as 0 minutes
        return 0
        

# Apply the extract_minutes function to the 'TimeColumn' and create a new column 'Minutes'
df['Minutes'] = df['time'].apply(extract_minutes)
```

```python
df.drop("time",axis=1,inplace=True)
```
