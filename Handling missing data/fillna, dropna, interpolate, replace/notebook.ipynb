<h1 style="background-color:orange;padding:10px 0 10px 10px;display:inline-block;">fillna()</h1>

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    "temperature" : 0,
    "wind_speed" : 0,
    "event" : "no event"
})
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill" , limit=1)
new_df
```

```python
new_df = df.fillna(method="ffill", axis="columns")
new_df
```

## Using Regrex to replace 

```python
df = pd.read_csv("temperature.csv")
df
```

## Now this data has lot of error

```python
new_df = df.replace({
    "temperature":'[a-zA-Z]',
    "wind_speed":'[a-zA-Z]',
    "event":'[0-9]'
},'',regex=True)
new_df
```

## Hurray ! Data is cleaned now
