<h1 style="background-color:orange;padding:10px 0 10px 10px;display:inline;">3) Sorting</h1>

```python
df.sort_values(by=["MILES*","STOP*"],ascending=[False,True])
```

<h1 style="background-color:orange;padding:10px 0 10px 10px;display:inline;">4) Conditionally adding columns</h1>

```python
import numpy as np
df["MILES_AVERAGE*"] = np.where(df["MILES*"]>=5,"Long Distance","Short Distance") 
df["New"] = np.array(20)
df.head(5)
```

```python
import numpy as np

conditions = [
    (df['MILES*'] >= 10),
    (df['MILES*'] < 10) & (df['MILES*'] > 5),
    (df['MILES*'] <= 5)
]

choices = ["Long Distance", "Medium Distance", "Short Distance"]

df["MILES_AVERAGE"] = np.select(conditions, choices, default="Unknown")
df.head(10)
```

<h1 style="background-color:orange;padding:10px 0 10px 10px;display:inline;">5) Grouping Data</h1>

```python
a = df.groupby("START*")["MILES*"].agg("mean")
b = a.sort_values(ascending=False)
b.head()
```

```python
df.groupby(["START*"])["MILES*"].agg(["mean","sum"])
```
